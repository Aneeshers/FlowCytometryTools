# Examples using the flow cytometry tools library

```python
%load_ext autoreload
%autoreload 2
from FlowCytometryTools import FCSample, FCPlate
```

Let's create a flow cytometry plate using all the fcs files located within the datadir path

```python


datadir = './FlowCytometryTools/tests/data/'
plate = FCPlate('test', datadir=datadir, shape=(8,12))
```

```python
figure(figsize=(10, 10))
plate.plot(['SC', 'Y2-A'], transform=('logicle', 'logicle'), plot2d_type='hist2d', grid_plot_kwargs={'well_ids':'A1'});

figure(figsize=(10, 10))
plate.plot(['B1-A', 'Y2-A'], transform=('logicle', 'logicle'), plot2d_type='hist2d', grid_plot_kwargs={'well_ids':'A1'});
```

```python
plate.apply(lambda x : x.ID, 'ID', applyto='sample')
plate.extracted

plate.wells_d['A1']

plate.apply(lambda x:x.ID, 'ID', applyto='sample', well_ids=['A1','B1'])
print plate.extracted
plate.apply(lambda x:x.datafile, 'file', applyto='sample')
plate.apply(lambda x:x.shape[0], 'counts', keepdata=True)
plate.get_well_metadata(['date', 'etim'])
print plate.extracted['file'].values
```

```python
x_channel = 'Y2-A'
y_channel = 'B1-A'

thisWell = plate.wells_d['A1']
thisWell.plot(x_channel, bins=100)
xlabel(x_channel)
ylabel('Counts')
plt.figure()
thisWell.plot([x_channel, y_channel], transform=('logicle', 'logicle'), plot2d_type='dot2d', marker='o', s=1)
xlabel(x_channel)
ylabel(y_channel)
figure()
thisWell.plot([x_channel, y_channel], transform=('logicle', 'logicle'), plot2d_type='hist2d')
xlabel(x_channel)
ylabel(y_channel)


```
