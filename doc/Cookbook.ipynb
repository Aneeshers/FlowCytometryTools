# Plotting

### Working with a single well

```python
well = plate['A3']
```

#### 1d histograms

<pre>
You will be doing all the plotting using the plot function. 
If you're plotting a single channel, the plot function will by default plot a histogram.

</pre>

```python
figure(figsize=(10, 10))
well.plot('B1-A', bins=100);
```

```python
figure(figsize=(10, 10))
well.plot('B1-A', bins=100, alpha=0.7, color='green', normed=1);
```

#### 2d histograms

```python
figure(figsize=(10, 10))
well.plot(['B1-A', 'Y2-A']);
```

## Transformations

<pre>
If you'll notice in the example above, the fluorescent population (high Y2-A) is squished against the non-fluorescent population (low Y2-A) values.
We can remedy this by asking the plot function to transform the data using an hlog transformation.
</pre>

```python
figure(figsize=(10, 10))
well.plot(['B1-A', 'Y2-A'], transform='hlog');
```

```python
f = figure(figsize=(14, 4))
subplots_adjust(wspace=0.4)
subplot(1, 3, 1)
well.plot(['B1-A', 'Y2-A']);

subplot(1, 3, 2)
well.plot(['B1-A', 'Y2-A'], transform=('hlog', 'hlog')); # This is equivalent to using transform='hlog'

subplot(1, 3, 3)
well.plot(['B1-A', 'Y2-A'], transform=(None, 'hlog'));

```

<pre>
This plot function accepts all the same arguments as does the matplotlib pcolormesh function.
For example, we can change the default colormap used for plotting.
</pre>

```python
well.plot(['B1-A', 'Y2-A'], transform=('hlog', 'hlog'), cmap=cm.Oranges);
```

<pre>
Also, instead of plotting histograms, we can choose to plot a scatter plot by setting the argument kind to 'scatter'.
In this case, the plot function will behave like the matplotlib scatter function and pass all additional arguments to scatter.
</pre>

```python
well.plot(['B1-A', 'Y2-A'], transform='hlog', kind='scatter', color='red', s=1, alpha=0.3)
```

## Sample Calculations

### Working with a single well

```python
well = plate.wells[1]['A'] # or one can use plate.wells_d['A1']
data = well.get_data()
print 'Channel names: {0}'.format(data.channels)
print 'Events, Channels: {0}, {1}'.format(data.shape[0], data.shape[1])
```

### Compute the total number of events

```python
def count_events(data):
    """ This functions counts the total number of events. """
    return data.shape[0]

print count_events(well.get_data())
```

#### Apply this calculation to the entire plate

```python
plate.apply(count_events, outputs='Total Event #', applyto='data')
```
